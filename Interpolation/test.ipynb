```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([0, 2, 4, 6, 7, 10, 12, 15, 17, 18, 20])
y = np.array([0, 2, 3, 4, 8, 10, 8, 12, 9, 11, 14])

# Fit a polynomial of degree n (choose degree based on complexity)
degree = len(x)-1
coefficients = np.polyfit(x, y, degree)

# Generate the polynomial equation
polynomial = np.poly1d(coefficients)

# Print the equation
print(f"Polynomial coefficients: {coefficients}")
print(f"Polynomial equation:\n{polynomial}")

# Plot the data points and the polynomial fit
x_fit = np.linspace(min(x), max(x), 500)
y_fit = polynomial(x_fit)

plt.scatter(x, y, color='red', label='Data points')
plt.plot(x_fit, y_fit, label=f'Polynomial fit (degree {degree})', color='blue')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fit')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = [0, 2, 4, 6, 7, 10, 12, 15, 17, 18, 20]
y = [0, 2, 3, 4, 8, 10, 8, 12, 9, 11, 14]

# Function to perform Lagrange interpolation
def lagrange_interpolation(x_points, y_points, x_val):
    n = len(x_points)
    result = 0
    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if i != j:
                term *= (x_val - x_points[j]) / (x_points[i] - x_points[j])
        result += term
    return result

# Generate interpolated values
x_fit = np.linspace(min(x), max(x), 500)
y_fit = [lagrange_interpolation(x, y, xi) for xi in x_fit]


# Plot the original data and the interpolation
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x_fit, y_fit, color='blue', label='Lagrange interpolation')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Interpolation')
plt.grid(True)
plt.show()
```
